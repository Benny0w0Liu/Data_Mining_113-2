```python
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload

import KNN
import PLA
import SVM

reload(KNN)
reload(PLA)
reload(SVM)
```

# experiment A

```python
# KNN
ax = plt.gca()
ax.set_ylim([0, 100])
x = np.array(list(range(1,200,2)))
print(len(x))
y=[]
for k in x:
    y.append(KNN.experiment("A",k))

plt.scatter(x[np.argmax(y)], y[np.argmax(y)], color='red', zorder=5)
plt.text(x[np.argmax(y)], y[np.argmax(y)], f'({x[np.argmax(y)]:.2f}, {y[np.argmax(y)]:.2f})', fontsize=12)
plt.plot(x, y, 'b-')
```
