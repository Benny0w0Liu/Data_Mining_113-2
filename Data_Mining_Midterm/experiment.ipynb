```python
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload

import KNN
import PLA
import SVM

reload(KNN)
reload(PLA)
reload(SVM)
```

# KNN

```python
# Experiment A
ax = plt.gca()
ax.set_ylim([0, 100])
x = np.array(list(range(1,200,2)))
y=[]
for k in x:
    y.append(KNN.experiment("A",k))

print(f'Highest accuracy(marked point) : k = {x[np.argmax(y)]:.2f} -> Accuracy = {y[np.argmax(y)]:.2f}%')
plt.scatter(x[np.argmax(y)], y[np.argmax(y)], color='red', zorder=5)
plt.xlabel('value of k')
plt.ylabel('accuracy')
plt.title(f'KNN - Experiment A', fontsize=12)
plt.plot(x, y, 'b-')
```
